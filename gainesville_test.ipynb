```python
def rename_columns(dataframe):
    """Rename all cols to lowercase using a zip() between both lists"""
    cols = dataframe.columns.tolist()
    lc_cols = [col.lower().replace(' ', '_') for col in cols]
    renamed = dict(zip(cols, lc_cols))
    dataframe = dataframe.rename(columns=renamed)
    return dataframe
```

```python
def dob_convert(dob):
  dob = str(dob)
  if dob == 'nan':
    pass
  format_1 = re.compile(r'\d\d-\w\w\w-\d\d')
  format_2 = re.compile(r'\d\d/\d\d/\d\d')

  if re.match(format_1, dob):
    print('form 1')
    print(dob)
    converted = datetime.strptime(dob,'%d-%b-%y')
    print(converted)
    return converted
  elif re.match(format_2, dob):
    print('form 2', dob)
    converted = datetime.strptime(dob, '%m/%d/%y')
    print(converted)
    return converted
  else:
    converted = ''
    print('NA')
    return converted
```

```python
df.birth_date = df.birth_date.apply(dob_convert)
```
